# **Get the Data**

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
sp = yf.Ticker("^GSPC").history(period="max")
sp.index = pd.to_datetime(sp.index, utc= True) # convert index to datetime
```

```python
today = datetime.now().strftime("%Y-%m-%d")
```

```python
sp.tail()
```

```python
sp.drop(columns=["Dividends", "Stock Splits"], inplace=True) # Inlpace = True to modify the dataframe
```

```python
sp.tail()
```

```python
sp["Tomorrow"] = sp["Close"].shift(-1)
```

```python
sp.tail()
```

```python
sp["Target"] = (sp["Tomorrow"] > sp["Close"]).astype(int) # 1 if the price goes up, 0 if it goes down
```

```python
sp.head()
```

```python
sp.drop(columns=["Tomorrow"], inplace=True)
```

```python
sp.count()
```

```python
sp.to_csv(f"S&P500-{today}.csv")
```
